```python
class User:
    def __init__(self, first_name, last_name, username, email):
        self.first_name = first_name
        self.last_name = last_name
        self.username = username
        self.email = email
        self.login_attempts = 0 

    def increment_login_attempts(self):
        self.login_attempts += 1


class Admin(User):
    def __init__(self, first_name, last_name, username, email):
        super().__init__(first_name, last_name, username, email)
        
        self.privileges = ["can add post", "can delete post", "can ban user"]

    def show_privileges(self):
        print(f"\n⚡ Admin '{self.username}' has the following privileges:")
        for privilege in self.privileges:
            print(f" 🔹 {privilege}")


admin1 = Admin("Zeeshan", "Khan", "ZeeshanAdminAI", "admin.zeeshan@example.com")

print("--- Checking Inheritance ---")
print(f"Full Name: {admin1.first_name} {admin1.last_name}")
print(f"Email:     {admin1.email}")

admin1.show_privileges()
```

```python
class SkillAcademy:
    academy_name = "Aptech Saylani AI Academy"
    location = "Karachi, Pakistan"
    website_url = "https://www.skillacademy.com"

    def __init__(self):
        pass

    def student_admission(self, name, course):
        print(f" Admission Confirmed! Welcome '{name}' to the '{course}' program at {self.academy_name}.")

    @classmethod
    def print_contact_info(cls):
        print(f" Academy Location: {cls.location}")
        print(f" Name:             {cls.academy_name}")

    @staticmethod
    def calculate_discount(fee, discount_percentage):
        discount_amount = (fee * discount_percentage) / 100
        final_fee = fee - discount_amount
        return final_fee


SkillAcademy.print_contact_info()

discounted_fee = SkillAcademy.calculate_discount(5000, 10)
print(f" Original Fee: 5000 | After 10% Discount: {discounted_fee}")


academy_object = SkillAcademy()
academy_object.student_admission(name="Zeeshan Khan", course="Data Science")
```
