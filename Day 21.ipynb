```python
class Car:
    def __init__(self, make, model, horsepower, v_type):
        self.make = make
        self.model = model
        self.engine = self.Engine(horsepower, v_type)

    def show_car_info(self):
        print(f" Car: {self.make} {self.model}")
        print(f" Inner Engine: {self.engine.v_type} ({self.engine.horsepower} HP)")

    class Engine:
        def __init__(self, horsepower, v_type):
            self.horsepower = horsepower
            self.v_type = v_type

car2 = Car(make="BYD", model="Seal", horsepower=530, v_type="Dual-Motor Electric")
car2.show_car_info()
```

```python
class Parent1:
    def show_identity(self):
        print(" Method running from: Parent1")

class Parent2:
    def show_identity(self):
        print(" Method running from: Parent2")

class Child(Parent1, Parent2):
    pass


obj1 = Child()
obj1.show_identity()

print(f" Actual MRO Order: {Child.__mro__}")
```

```python
class ChildNew(Parent2, Parent1):
    pass

obj2 = ChildNew()
obj2.show_identity()

print(f" New MRO Order: {ChildNew.__mro__}")
```
