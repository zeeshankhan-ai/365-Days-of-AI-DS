```python

while True:

    response = input("Aapko programming kyun pasand hai? ")

    if response.lower().strip() == 'quit':
        print("\n Polling khatam! Data record kar liya gaya hai. Shukriya!")
        break 
        
    with open("programming_polls.txt", "a") as file:
        file.write(response + "\n")
        print(" Response saved successfully!")
```
