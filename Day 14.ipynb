```python

n = int(input("Enter a number (n): "))

odd_cubes = {}

for num in range(1, n + 1):
    
    if num % 2 != 0:
        odd_cubes[num] = num ** 3
        
    else:
        continue

        
print(odd_cubes)
```

```python
user_input = input("Enter words separated by commas")
words_list = user_input.split(",")
filtered_words = []

for word in words_list:
    clean_word = word.strip()
    
    if len(clean_word) > 4:
        filtered_words.append(clean_word)

filtered_words.sort()
print(filtered_words)
```

```python
def modify_names_matrix(names_list):

    updated_list = []
    
    for name in names_list:
        
        if name.startswith("S"):
            new_name = name + " Bhai"
            updated_list.append(new_name)
            
        elif name.endswith("A") or name.endswith("a"):
            new_name = name + " Khan"
            updated_list.append(new_name)
            
        else:
            new_name = name + " Jan"
            updated_list.append(new_name)
            
    return updated_list


friends = ["Shahzad", "Mustafa", "Sami", "Ali", "Faraz"]

modified_friends = modify_names_matrix(friends)

print(friends)

print(modified_friends)
```
