```python
std = [ {
    "student_name" : "zeeshan",
    "marks" : 90
},
      {
    "student_name" : "fazal",
    "marks" : 77
},
      {
    "student_name" : "faraz",
    "marks" : 26
},
      {
    "student_name" : "shahzad",
    "marks" : 65
}]


for student in std:
    name = student["student_name"]
    value = student["marks"]

    
    if (value>= 70) and (value < 80):
        print (f"{value} u are A grade ")
    elif(value>= 90) and (value <= 100):
        print (f"{value} u are A+1 grade ")
    elif(value >= 80) and (value < 90):
        print (f"{value} u are A+ grade ")
    elif(value>= 60) and (value < 70):
        print (f"{value} u are B grade ")
    elif(value>= 50) and (value < 60):
        print (f"{value} u are c grade ")
    else:
        print(f"Exam checking stopped because {name} failed.")
        break
```

```python

inventory = [
    {"item_name": "Calculator", "price": 1200, "inventory": 10},
    {"item_name": "Mouse", "price": 800, "inventory": 15},
    {"item_name": "Keyboard", "price": 1500, "inventory": 8}
]


for i in inventory:
    print(f"Product: {i['item_name']} | Price: Rs.{i['price']} | Stock Left: {i['inventory']}")


choice = input("Enter the item name you want to buy: ")
quantity = int(input("How many do you want to buy? "))

item_found = False

for i in inventory:
    if i["item_name"].lower() == choice.lower():
        item_found = True
        
        if i["inventory"] >= quantity:
            
 
            total_bill = i["price"] * quantity
            print(f"\nTotal bill is: Rs.{total_bill}")
            print(f" Please pay Rs.{total_bill} ")
            
       
            i["inventory"] = i["inventory"] - quantity
            print("Payment Successful! Inventory Updated.")
            
        else:
            print(f" Sorry, we don't have enough stock! Only {i['inventory']} left.")
        break

if not item_found:
    print(" Item not found in our shop!")


print("\n Updated Shop Inventory Stock ")
for i in inventory:
    print(f"Product: {i['item_name']} | Stock Remaining: {i['inventory']}")
```
