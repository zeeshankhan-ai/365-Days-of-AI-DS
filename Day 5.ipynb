Quadratic Equation (Discriminant)

```python
a = float(input("Enter value of a x"))
b = float(input("Enter value of b x^2" ))
c = float(input("Enter value of c"))

d = (b ** 2) - (4*a*c)

if d == 0:
    print("Equal Roots")
elif d > 0:
    print("Real Roots")
else:
    print("Imaginary")
    
```
