Ticket Price Calculator (Age Based)


```python
age = int(input("enter your age"))

if age <= 5:
    print("Ticket is Free!")
elif (age > 5) and (age < 18):
    print("Half Ticket: Rs. 500")
else:
    print("Full Ticket: Rs. 1000")
```
