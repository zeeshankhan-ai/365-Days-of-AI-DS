```python

salary = int(input("enter your salary"))

def gym_deduction_calculator(basic_salary):
    if basic_salary > 50000:
        return 2000
    else:
        return 1000

def insurance_calculator(basic_salary):
    return basic_salary * 0.02

def disburse_salary(basic_salary):
 
    gym_fee = gym_deduction_calculator(basic_salary)
    insurance_fee = insurance_calculator(basic_salary)
    
    total_deductions = gym_fee + insurance_fee

    final_take_home = basic_salary - total_deductions
    
    print(f"Gross Salary:     Rs.{basic_salary}")
    print(f"Gym Deduction:    Rs.{gym_fee}")
    print(f"Insurance (2%):   Rs.{insurance_fee}")
    print(f"Total Deductions: Rs.{total_deductions}")
    print(f" Net Salary:    Rs.{final_take_home}")

disburse_salary(salary)
```

```python
def overtime_calculator(hours_worked, rate_per_hour):
    if hours_worked > 160:
        extra_hours = hours_worked - 160
        overtime_bonus = extra_hours * (rate_per_hour * 1.5)
        return overtime_bonus
    else:
        return 0 

def disburse_salary(basic_salary, hours_worked, rate_per_hour):
    overtime_bonus = overtime_calculator(hours_worked, rate_per_hour)
    
    total_earnings = basic_salary + overtime_bonus
    
    print(f"Basic Salary:    Rs.{basic_salary}")
    print(f"Hours Worked:    {hours_worked} hrs (Normal: 160 hrs)")
    print(f"Overtime Bonus:  Rs.{overtime_bonus}")
    print(f" Total Payout:  Rs.{total_earnings}")



disburse_salary(basic_salary=60000, hours_worked=180, rate_per_hour=500)

# Test Case 2: Bande ne normal 150 ghante kaam kiya (No overtime)
disburse_salary(basic_salary=60000, hours_worked=150, rate_per_hour=500)
```
