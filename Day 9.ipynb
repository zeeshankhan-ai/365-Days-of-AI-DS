Filter and Capitalize Names (Basic)

```python
students_list = ["farhad", "amir", "mohit", "ali", "rehan"]

for name in students_list:
    if name.startswith("a"):
        print(name.upper())
    else:
        print (name)
```

Course Wise Fee Counter (Intermediate)

```python
students_data = [
    {"name": "Farhan", "course": "Data Science", "fee": 5000},
    {"name": "Salman", "course": "Python", "fee": 4000},
    {"name": "Ali", "course": "Data Science", "fee": 6000},
    {"name": "Asif", "course": "Python", "fee": 4000},
    {"name": "Sana", "course": "Data Science", "fee": 5500}
]

total_fee = 0

for student in students_data:

    if student["course"] == "Data Science":
        total_fee += student["fee"]

print(f"Total fee collected from Data Science students: {total_fee}")
```

Dynamic Data Science Analytics (Advanced)

```python
customers = [
    {"name": "Zane", "spend": 120.50, "membership": "Regular"},
    {"name": "Ayesha", "spend": 350.00, "membership": "Premium"},
    {"name": "Kamran", "spend": 80.00, "membership": "Regular"},
    {"name": "Bilal", "spend": 450.00, "membership": "Premium"}
]


total_spend = 0
for customer in customers:
    total_spend += customer["spend"]

    if customer["membership"] == "Premium":
        print(customer["name"])

averagespend = total_spend / len(customers)
print(f"Average spend of customers is: {averagespend}")
```
