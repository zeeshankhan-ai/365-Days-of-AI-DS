```python

def calculate_tax(amount):
    
    if amount >= 100000:
        tax_percentage = 30
        tax_amount = amount * 0.30  

    elif amount >= 50000:
        tax_percentage = 15
        tax_amount = amount * 0.15  
        
    else:
        tax_percentage = 0
        tax_amount = 0  
        
    print(f"Salary: Rs.{amount} | Tax Bracket: {tax_percentage}% | Tax to Pay: Rs.{tax_amount}")


salary = int(input("enter your monthly income"))

calculate_tax(salary)  
  
```
