```python
class Vehicle:
    def __init__(self, name):
        self.name = name

    def start_engine(self):
        print(f" {self.name}: Engine started with fuel.")

class ElectricBike(Vehicle):
    def start_engine(self):
        print(f" {self.name}: No engine! Started quietly with Battery power.")

normal_car = Vehicle("Honda Civic")
normal_car.start_engine() 

my_ebike = ElectricBike("Super Star E-Bike")
my_ebike.start_engine()
```

```python
from abc import ABC, abstractmethod

# 1. Abstract Parent Class (Blueprint)
class BankAccount(ABC):
    def __init__(self, balance):
        self.balance = balance

    @abstractmethod
    def calculate_interest(self):
        pass 

class SavingsAccount(BankAccount):
    def calculate_interest(self):
        return self.balance * 0.05  

class CurrentAccount(BankAccount):
    def calculate_interest(self):
        return 0  


savings = SavingsAccount(100000)
print(f" Savings Account Interest: {savings.calculate_interest()} PKR")

current = CurrentAccount(100000)
print(f" Current Account Interest: {current.calculate_interest()} PKR")
```
