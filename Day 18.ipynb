```python

class Student:

    def __init__(self, roll_number, name, course, age):
        self.roll_number = roll_number
        self.name = name
        self.course = course
        self.age = age



student1 = Student(roll_number="DS-101", name="Zeeshan Khan", course="Data Science", age=22)
student2 = Student(roll_number="AI-202", name="Asad Ali", course="Artificial Intelligence", age=24)




print(f"Name:   {student1.name}")
print(f"Course: {student1.course}")

print(f"Name:   {student2.name}")
print(f"Course: {student2.course}")

print(f" Student 1 Memory Address: {hex(id(student1))}")
print(f" Student 2 Memory Address: {hex(id(student2))}")
```

```python

class Car:
    def __init__(self, brand, model, engine_cc, color):
        self.brand = brand
        self.model = model
        self.engine_cc = engine_cc
        self.color = color
        
    def start_car(self):
        print(f" Car {self.brand} {self.model} has started successfully!")


my_car = Car(brand="BYD", model="Seal", engine_cc="Electric (N/A)", color="Cosmos Black")

my_car.start_car()
```
