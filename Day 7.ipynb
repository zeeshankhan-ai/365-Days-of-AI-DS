```python
profile = {
    "name": "Formo",
    "platform": "Meta",
    "adspend": 5000,      
    "conversions": 125,    
    "isprofitable": True   }

spend = profile["adspend"]
conversions = profile["conversions"]
cpa = spend / conversions

print(f"Name: {profile['name']}")
print(f"Platform: {profile['platform']}")
print(f"Total Spend: {spend}")
print(f"Total Conversions: {conversions}")
print(f"Cost Per Acquisition CPA: {cpa:.2f} per conversion ")
```
